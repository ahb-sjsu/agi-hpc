````python
import subprocess


def check(cmd, label):
    try:
        r = subprocess.run(cmd, capture_output=True, text=True, timeout=5)
        if r.returncode == 0:
            first_line = r.stdout.strip().split('\n')[0]
            print(f'✅ {label}: {first_line}')
            return True
        print(f'❌ {label}: returned {r.returncode}')
        return False
    except FileNotFoundError:
        print(f'❌ {label}: not installed')
        return False
    except Exception as e:
        print(f'❌ {label}: {e}')
        return False


check(['git', '--version'], 'git')
check(['gh', '--version'], 'gh CLI')  # install from cli.github.com if missing
check(['docker', '--version'], 'docker')  # nice-to-have for Quest 5
````

## 🎯 Quest 2 — The Bootstrap

**[Issue #67](https://github.com/ahb-sjsu/agi-hpc/issues/67)** · ⏱ 1 day · 🔧 Scripting

### Mission

Write a script that takes a fresh laptop from `git clone` to *"all tests pass"* in one command. When the next contributor joins and runs it, they should be running smoke tests 5 minutes later.

### Your deliverable

- `scripts/bootstrap.sh` **or** `scripts/bootstrap.ps1` — pick one based on what you run
- Runs `pytest tests/test_basic.py` at the end as verification
- Exits non-zero on any failure with a clear message
- `scripts/README.md` explaining when to rerun it (first setup, after dep changes, etc.)

### Skeleton (bash)

```bash
#!/usr/bin/env bash
set -euo pipefail

# Colors for humans
RED='\033[0;31m'; GREEN='\033[0;32m'; YELLOW='\033[1;33m'; NC='\033[0m'
info()  { echo -e "${YELLOW}▶ $*${NC}"; }
ok()    { echo -e "${GREEN}✓ $*${NC}"; }
die()   { echo -e "${RED}✗ $*${NC}"; exit 1; }

# 1. Check Python version
# 2. Create venv if missing
# 3. Activate venv
# 4. Install deps (editable, dev extras)
# 5. Run smoke tests
# 6. Print success banner
```

### 💡 Hints

- `set -euo pipefail` at the top. Read about why. You'll thank yourself.
- Detect an existing venv — don't blow it away silently. Prompt or accept a `--force` flag.
- `python3 -m venv` ships with Python. No need for virtualenv.
- After creating the venv, `source .venv/bin/activate` in the *script* won't affect the caller's shell. Print instructions for the caller to activate.
- Colored output with plain ANSI codes (no deps) looks great in every terminal.

### 🐚 PowerShell version?

```powershell
$ErrorActionPreference = 'Stop'
function Info($m) { Write-Host "▶ $m" -ForegroundColor Yellow }
function Ok($m)   { Write-Host "✓ $m" -ForegroundColor Green }
function Die($m)  { Write-Host "✗ $m" -ForegroundColor Red; exit 1 }
```

### ⚠️ Gotchas

- **Windows line endings** will break bash scripts. Add `* text=auto eol=lf` to `.gitattributes`.
- **Git Bash vs WSL** are different. State which you target in the README.
- **Don't assume `python` exists** — on many systems it's `python3`.

### 🌟 Stretch

- Add `--python X.Y` flag to force a specific Python version.
- Cache pip downloads across runs (`pip install --cache-dir .pip_cache`).
- Detect Apple Silicon vs Intel and warn about known gotchas.

### ✅ Done when

- Deleting `.venv/` and rerunning the script gets you back to green tests in under 5 minutes on your machine
- Another student on Discord runs it on their laptop and it works first try

## 🎯 Quest 3 — The Doctor

**[Issue #68](https://github.com/ahb-sjsu/agi-hpc/issues/68)** · ⏱ 2–3 days · 🩺 Tool-building

### Mission

Build `python -m agi_hpc.doctor` — a diagnostic tool that checks a dev environment and reports pass/fail on each check with actionable fix suggestions. Think `brew doctor` or `gh auth status`.

The point is not just "it runs checks" — the point is that when something is broken, the *user knows exactly what to do next*. A message like `port 8000 is in use` is useless. `port 8000 is in use (kill process: lsof -ti:8000 | xargs kill)` is a Doctor.

### Your deliverable

- A module at `src/agi_hpc/doctor/__init__.py` (or wherever fits)
- `python -m agi_hpc.doctor` runs all checks
- Each check prints: name · status (✅/❌/⚠️) · one-line reason · fix suggestion if failing
- Exits `0` on all-pass, `1` otherwise
- At least 6 checks (list below)

### Check ideas (pick ≥6)

| # | What to check | Good fix suggestion |
|---|---|---|
| 1 | Python >= 3.10 | `Install Python 3.12 from python.org` |
| 2 | venv activated | `source .venv/bin/activate` |
| 3 | Key packages importable | `pip install -e '.[dev]'` |
| 4 | Git is configured | `git config --global user.name 'Your Name'` |
| 5 | Disk space > 10 GB | `Free up space, heavy builds need room` |
| 6 | Common ports free | `lsof -ti:8000 \| xargs kill` |
| 7 | `.env` exists (if applicable) | `cp .env.example .env` |
| 8 | Docker daemon running | `Start Docker Desktop` |
| 9 | `gh` CLI authenticated | `gh auth login` |

### Code skeleton

```python
from dataclasses import dataclass
from collections.abc import Callable

@dataclass
class CheckResult:
    name: str
    passed: bool
    message: str
    suggestion: str = ''
    warning: bool = False

Check = Callable[[], CheckResult]

def check_python_version() -> CheckResult:
    import sys
    ok = sys.version_info >= (3, 10)
    return CheckResult(
        name='Python version',
        passed=ok,
        message=f'{sys.version.split()[0]}',
        suggestion='' if ok else 'Install Python 3.10+ from python.org',
    )

CHECKS: list[Check] = [check_python_version, ...]

def main() -> int:
    results = [c() for c in CHECKS]
    # print report, return 0 or 1
    ...
```

### 💡 Hints

- The `rich` library makes colored output effortless: `from rich.console import Console; Console().print('[green]✓[/green]')`. If you want zero deps, plain ANSI also works.
- `shutil.which('git')` tells you if a command is on PATH. Don't shell out unless you need the command's output.
- Every check should be a **pure function** returning a `CheckResult`. That way testing each check is trivial.
- Write a test per check. Pass a `monkeypatch` fixture to simulate missing tools.
- If you add `--verbose`, show the raw command output on failing checks.

### ⚠️ Gotchas

- **Don't execute untrusted input** inside your checks. Ever. Even for fix suggestions.
- **Cross-platform** — `lsof` doesn't exist on Windows. Use `psutil` if you want portability.
- **Subprocess timeouts** — always pass `timeout=5` to `subprocess.run`. Hung commands ruin the UX.

### 🌟 Stretch

- `--json` output mode for CI consumption.
- `--fix` mode that auto-fixes fixable problems (create venv, install deps, etc.).
- Plugin system: let external code register new checks via entry points.

### ✅ Done when

- Running it on a clean machine prints a useful report
- Running it in CI as a sanity gate catches at least one real problem you introduced

## 🎯 Quest 5 — The Container

**[Issue #70](https://github.com/ahb-sjsu/agi-hpc/issues/70)** · ⏱ 2–3 days · 🐳 Docker

### Mission

Pick one small, self-contained service in the codebase, write a production-quality Dockerfile, and show the before/after. When you're done, we'll have a repeatable pattern for every other service.

### Your deliverable

- `docker/<service>.Dockerfile` — multi-stage build (builder + runtime), non-root user, pinned base image
- `.dockerignore` that excludes the obvious junk (`__pycache__`, `.venv`, `.git`, etc.)
- `docker/<service>-README.md` — build, run, verify instructions
- A GitHub Actions job that builds the image on PR (don't push to a registry yet)
- **Image size under 500 MB** (this forces good layer hygiene)

### Dockerfile pattern (multi-stage)

```dockerfile
# syntax=docker/dockerfile:1.7

# ---- builder stage ----
FROM python:3.12-slim AS builder
ENV PIP_NO_CACHE_DIR=1 PYTHONDONTWRITEBYTECODE=1
WORKDIR /app
COPY pyproject.toml ./
# copy only what's needed to resolve deps
RUN pip install --prefix=/install .

# ---- runtime stage ----
FROM python:3.12-slim AS runtime
ENV PYTHONDONTWRITEBYTECODE=1 PYTHONUNBUFFERED=1
RUN useradd --create-home --shell /bin/bash appuser
COPY --from=builder /install /usr/local
COPY src/ /app/src/
WORKDIR /app
USER appuser
HEALTHCHECK --interval=30s --timeout=3s CMD python -c 'import agi_hpc' || exit 1
CMD ["python", "-m", "agi_hpc.service"]
```

### 💡 Hints

- `python:3.12-slim` over `python:3.12` — saves ~100 MB.
- `--no-cache-dir` on pip saves another ~100 MB.
- A missing `.dockerignore` is the #1 cause of bloated builds. Check your `docker build` output — if it says "transferring context: 500 MB", something is wrong.
- `docker history <image>` shows per-layer sizes. Fat layers are where to look.
- Non-root user (`USER appuser`) is a security basic. Some CVE scanners will fail the image if you don't.

### Size check

```bash
docker build -t mysvc -f docker/mysvc.Dockerfile .
docker images mysvc --format "{{.Size}}"
# 450MB — under target ✓
```

### ⚠️ Gotchas

- If your service reads/writes files, be explicit about `VOLUME` and `WORKDIR`. Implicit is painful.
- **Tag, don't use `:latest`** for the GHA build — use the commit SHA.
- Healthchecks fail silently by default. Use `docker inspect --format='{{.State.Health}}' <container>` to debug.

### 🌟 Stretch

- `docker-compose.yml` that starts the service alongside postgres for integration testing.
- Image signing with `cosign` (real production practice).
- Multi-arch build (`linux/amd64,linux/arm64`) using `buildx`.

### ✅ Done when

- `docker build` succeeds in CI
- `docker run` actually runs the service and healthcheck passes
- Image is under 500 MB
- Another student can follow your README and reproduce your build

## 🆘 When and How to Ask for Help

### When to ask

- 🟢 **Immediately:** environment/setup issues (someone else can unblock you in 2 minutes)
- 🟢 **After 30 min:** "I can't figure out why X"
- 🟢 **After 1 hour:** any time you're circling
- 🟡 **After trying:** design questions, architecture choices
- 🔴 **Don't wait longer than half a day.** Seriously.

### How to ask well (the XY problem)

Bad: *"How do I check if a port is in use?"*

Good: *"I'm trying to check if port 8000 is available for the Doctor tool. I tried `lsof -ti:8000` but it's not on Windows. Any ideas?"*

The difference: the second version tells us your **goal** (port check for Doctor), what you **tried** (lsof), and **why it didn't work** (not on Windows). Now we can give you a useful answer on the first try.

### Where to ask

| Question type | Where |
|---|---|
| Environment / setup | Discord #setup channel |
| Quest-specific blocker | Comment on the Quest's GitHub issue |
| "Is this the right approach?" | Draft PR + ask for early review |
| "I broke something" | Discord. Fast. |
| "How does X work conceptually?" | GitHub Discussion |

## 📎 Cheatsheet

### Git you'll use every day

```bash
git status                           # what am I looking at
git diff                             # what changed, not staged
git diff --staged                    # what changed, staged
git add -p                           # stage interactively (powerful)
git commit -m 'msg'                  # commit staged changes
git log --oneline -20                # last 20 commits
git log -p <file>                    # history of one file, with diffs
git blame <file>                     # who changed each line when
git checkout -- <file>               # discard unstaged changes (careful!)
git stash / git stash pop            # tuck changes aside temporarily
```

### Python hygiene

```bash
python -m venv .venv                # create venv
```
